```python
filepath3 = "../../Scraping/IYK- ETL Project/scrape_indeed_combined.csv"
salary = pd.read_csv(filepath3)
salary.head()
```

```python
salary["location"]=salary["location"].fillna("Remote")
salary["salary"]=salary["salary"].fillna("Unknown")
```

```python
salary["city"]=""
```

```python
for index, row in salary.iterrows():
    salary.loc[index, 'city'] = salary.loc[index,'location'].split(",")[0]
salary
```

```python
salary.to_csv('../Resources/company_salary.csv', index=False)
```
